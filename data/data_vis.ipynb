```python
import os
import pandas as pd
import numpy as np

data_dir = "wavelength/"

# Step 2: Get all .txt files in that folder
txt_files = [f for f in os.listdir(data_dir) if f.endswith(".txt")]

# Step 3: Dictionary to hold DataFrames per file
dataframes = {}

# Step 4: Read each wl1 and wl2 text file
for filename in txt_files:
    full_path = os.path.join(data_dir, filename)
    base_name = filename.split("_")[0]  # grabs just '001'
    suffix = "wl1" if "wl1" in filename else "wl2"

    # Parse all whitespace-separated numbers into float array
    with open(full_path, 'r') as file:
        values = [float(x) for x in file.read().split()]
    
    rows = len(values) // 120
    reshaped = np.array(values[:rows * 120]).reshape(rows, 120)
    
    df = pd.DataFrame(reshaped, columns=[f"{suffix}_ch{i+1}" for i in range(120)])
    dataframes[f"{base_name}_{suffix}"] = df

# Step 5: Combine wl1 and wl2 for each subject
combined_data = {}
for subject_id in set(k.replace("_wl1", "").replace("_wl2", "") for k in dataframes.keys()):
    if f"{subject_id}_wl1" in dataframes and f"{subject_id}_wl2" in dataframes:
        combined_df = pd.concat([dataframes[f"{subject_id}_wl1"], dataframes[f"{subject_id}_wl2"]], axis=1)
        combined_data[subject_id] = combined_df

# Step 6: Preview the combined dataframe for subject 001
combined_data["001"].head()
```

```python
df = combined_data["001"].copy()

# Constants
dpf = 6.0          # Differential pathlength factor
distance = 3.0     # cm
L = dpf * distance # Optical pathlength
ε_HbO = np.array([1486, 2526])  # 760nm, 850nm
ε_HbR = np.array([3843, 1798])

# Get wl1 and wl2 data (assumed ordered ch1 to ch120)
wl1 = df[[col for col in df.columns if "wl1" in col]].to_numpy()
wl2 = df[[col for col in df.columns if "wl2" in col]].to_numpy()

# Take log (change in intensity relative to baseline = first time point)
I0_wl1 = wl1[0, :]
I0_wl2 = wl2[0, :]
ΔA_wl1 = -np.log(wl1 / I0_wl1)
ΔA_wl2 = -np.log(wl2 / I0_wl2)

# Solve MBLL for each channel (per time step)
ΔA = np.stack([ΔA_wl1, ΔA_wl2], axis=2)  # shape (n_timepoints, 120, 2)

# Invert extinction matrix
E = np.array([ε_HbO, ε_HbR]).T  # shape (2,2)
E_inv = np.linalg.inv(E)

# Compute [HbO], [HbR] changes
n_timepoints = ΔA.shape[0]
HbO_changes = np.zeros((n_timepoints, 120))
HbR_changes = np.zeros((n_timepoints, 120))

for t in range(n_timepoints):
    result = E_inv @ ΔA[t].T / L
    HbO_changes[t] = result[0]
    HbR_changes[t] = result[1]

# Build output DataFrame
df_hbo = pd.DataFrame(HbO_changes, columns=[f"HbO_ch{i+1}" for i in range(120)])
df_hbr = pd.DataFrame(HbR_changes, columns=[f"HbR_ch{i+1}" for i in range(120)])
df_out = pd.concat([df_hbo, df_hbr], axis=1)

# Inspect result
display(df_out.head())

# Save to JSON
df_out.to_json("subject_001_hemodynamics.json", orient="records", indent=2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import os

def load_data(subject, condition, wavelength, folder='wavelength'):
    file_name = f"{subject}_{condition}_{wavelength}.txt"
    file_path = os.path.join(folder, file_name)
    return np.loadtxt(file_path)  # 2D: (timepoints, channels)

def compute_delta_concentration(I_760, I_850):
    """
    Compute relative HbO and HbR per channel using differential MBLL.
    Returns: HbO, HbR → shape (timepoints, channels)
    """
    I_760[I_760 <= 0] = 1e-10
    I_850[I_850 <= 0] = 1e-10

    # Convert to OD
    OD_760 = -np.log(I_760 / np.mean(I_760[:100, :], axis=0))
    OD_850 = -np.log(I_850 / np.mean(I_850[:100, :], axis=0))

    # Extinction coefficients
    e = np.array([[0.15, 1.486],   # 760 nm
                  [0.586, 0.174]]) # 850 nm

    num_timepoints, num_channels = OD_760.shape
    HbO = np.zeros((num_timepoints, num_channels))
    HbR = np.zeros((num_timepoints, num_channels))

    for t in range(num_timepoints):
        for c in range(num_channels):
            A = np.array([OD_760[t, c], OD_850[t, c]])
            conc, _, _, _ = np.linalg.lstsq(e, A, rcond=None)
            HbO[t, c], HbR[t, c] = conc

    return HbO, HbR

# Load subject 001
I_before_w1 = load_data('001', 'before', 'wl1')
I_before_w2 = load_data('001', 'before', 'wl2')
I_after_w1  = load_data('001', 'after',  'wl1')
I_after_w2  = load_data('001', 'after',  'wl2')

HbO_before, HbR_before = compute_delta_concentration(I_before_w1, I_before_w2)
HbO_after,  HbR_after  = compute_delta_concentration(I_after_w1,  I_after_w2)

# Plot example for first channel
channel = 0

plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(HbO_before[:, channel], label='Before SD', color='blue')
plt.plot(HbO_after[:, channel], label='After SD', color='red')
plt.title(f'ΔHbO (Oxyhemoglobin) - Channel {channel}')
plt.ylabel('Concentration (a.u.)')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(HbR_before[:, channel], label='Before SD', color='blue')
plt.plot(HbR_after[:, channel], label='After SD', color='red')
plt.title(f'ΔHbR (Deoxyhemoglobin) - Channel {channel}')
plt.ylabel('Concentration (a.u.)')
plt.xlabel('Time (samples)')
plt.legend()

plt.tight_layout()
plt.show()
```
