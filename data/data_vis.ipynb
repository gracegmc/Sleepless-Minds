```python
import os
import pandas as pd
import numpy as np

data_dir = "wavelength/"

# Step 2: Get all .txt files in that folder
txt_files = [f for f in os.listdir(data_dir) if f.endswith(".txt")]

# Step 3: Dictionary to hold DataFrames per file
dataframes = {}

# Step 4: Read each wl1 and wl2 text file
for filename in txt_files:
    full_path = os.path.join(data_dir, filename)
    base_name = filename.split("_")[0]  # grabs just '001'
    suffix = "wl1" if "wl1" in filename else "wl2"

    # Parse all whitespace-separated numbers into float array
    with open(full_path, 'r') as file:
        values = [float(x) for x in file.read().split()]
    
    rows = len(values) // 120
    reshaped = np.array(values[:rows * 120]).reshape(rows, 120)
    
    df = pd.DataFrame(reshaped, columns=[f"{suffix}_ch{i+1}" for i in range(120)])
    dataframes[f"{base_name}_{suffix}"] = df

# Step 5: Combine wl1 and wl2 for each subject
combined_data = {}
for subject_id in set(k.replace("_wl1", "").replace("_wl2", "") for k in dataframes.keys()):
    if f"{subject_id}_wl1" in dataframes and f"{subject_id}_wl2" in dataframes:
        combined_df = pd.concat([dataframes[f"{subject_id}_wl1"], dataframes[f"{subject_id}_wl2"]], axis=1)
        combined_data[subject_id] = combined_df

# Step 6: Preview the combined dataframe for subject 001
combined_data["001"].head()
```
